```python
import glob
import operator

hamfiles = []
spamfiles = []
ham_count = 0
spam_count = 0

for file in glob.glob("./enron1/ham/*.txt"):
    hamfiles.append(file)
    ham_count += 1
    
for file in glob.glob("./enron1/spam/*.txt"):
    spamfiles.append(file)
    spam_count += 1
    
print('ham count', ham_count)
print('spam count', spam_count)

```

```python
hamsets = []
for hamfile in hamfiles:
    txt = open(hamfile, "r", encoding="latin-1")
    hamsets.append(set(txt.read().split()))
    
spamsets = []

for spamfile in spamfiles:
    txt = open(spamfile, "r", encoding="latin-1")
    spamsets.append(set(txt.read().split()))
    ham_hash = {}
all_hash = {}
for hamset in hamsets:
    for word in hamset:
        if word in all_hash:
            all_hash[word] += 1
        else:
            all_hash[word] = 1
            
        if word in ham_hash:
            ham_hash[word] += 1
        else:
            ham_hash[word] = 1
            
spam_hash = {}
for spamset in spamsets:
    for word in spamset:
        if word in spam_hash:
            spam_hash[word] += 1
        else:
            spam_hash[word] = 1
            
        if word in all_hash:
            all_hash[word] += 1
        else:
            all_hash[word] = 1
print(ham_hash)
```

```python
base_ratio = ham_count / spam_count
odds_ratio = {}
for word, ham_count in ham_hash.items():
    spam_count = spam_hash.get(word) or 0.5
    odds_ratio[word] = base_ratio /(ham_count / spam_count)
    
for word, spam_count in spam_hash.items():
    if odds_ratio.get(word):
        continue
    odds_ratio[word] = base_ratio / (spam_count / 0.5) 

    
print(sorted(odds_ratio.items(), key=operator.itemgetter(1), reverse=True))
    
```
